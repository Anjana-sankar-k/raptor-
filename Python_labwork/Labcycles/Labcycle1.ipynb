```python
#PROGRAM 1

number=int(input("Enter a four digit number: "))
reverse=0

ones=number%10
tens=(number//10)%10
hundreds=(number//100)%10
thousands=number//1000
sum =ones+tens+hundreds+thousands
print("The sum of digits: ",sum)

while(number!=0):
  n=number%10
  number=number//10
  reverse=(reverse*10)+n
print("the reverse of the number is: ",reverse)

difference= (thousands*tens)-(ones*hundreds)
print("The difference is: ",difference)


```

```python
#PROGRAM 2
def get():
  print("Enter the sides of triangle")
  a=int(input("Enter the length of first side: "))
  b=int(input("Enter the length of second side: "))
  c=int(input("Enter the length of third side: "))
  return a,b,c

def area(a,b,c):
  s=(a+b+c)//2
  area= (s*(s-a)*(s-b)*(s-c))**(1/2)
  return area

a,b,c=get()
area1=area(a,b,c)
d,e,f=get()
area2=area(d,e,f)
total=area1+area2
per1=(area1/total)*100
per2=(area2/total)*100
print("The total area is: ",total)
print("The contribution of first triangle is: ",per1)
print("The contribution of second triangle is: ",per2)

```

```python
#PROGRAM 4
def check_number(n):
  for i in range(1,101):
    sum=0
    while (n>0):
      remainder=n%10
      sum+=(remainder**2)
      n=n//10
    if (sum==1):
      return True
    elif(n<1):
      n=sum
    else:
      return False


def In_range(lower,upper):
  count=0
  for i in range(lower,upper):
    if check_number(i)==True:
      count+=1
      print(i)
  if count==0:
    print("There are no happy numbers in this range.")


def print_happyno(x):
  count=0
  i=1
  while(count!=x):
    if check_number(i)==True:
      print(i)
      count+=1
    i+=1

    if count==x:
      break

n=int(input("Enter the number you want to check: "))
if (check_number(n)==True):
  print("It's a happy number.")
else:
  print("It's a sad number.")

lower=int(input("Enter the lower limit of range: "))
upper=int(input("Enter the upper limit of range: "))
In_range(lower,upper)

input_no=int(input("Enter the number of happy numbers you want printed: "))
print_happyno(input_no)
```

```python
#PROGRAM 5

def print_substrings(string):
    number = len(string)
    for i in range(number):
        for j in range(i+1,number+1):
            print(string[i:j])

def print_substrings_len_k(string, k):
    number = len(string)
    if k<=0 or k>number:
        print("Invalid length.")
        return

    for i in range(number-k+1):
        substring = string[i:i +k]
        print(substring)

def print_substrings_len_k_distinct(string, k):
    number = len(string)
    if k<=0 or k>number:
        print("Invalid length.")
        return

    for i in range(number-k+1):
        substring = string[i:i +k]
        if len(set(substring)) == k:
            print(substring)

def print_max_length_substrings_distinct(string, number):
    max_length = 0
    max_substrings = []

    for i in range(len(string)):
        chars_set = set()
        current_length = 0
        current_substring = ""

        for j in range(i, len(string)):
            if string[j] not in chars_set:
                chars_set.add(string[j])
                current_length += 1
                current_substring += string[j]
            else:
                break

        if current_length > max_length:
            max_length = current_length
            max_substrings = [current_substring]
        elif current_length == max_length:
            max_substrings.append(current_substring)

    print("Maximum length substrings with {} distinct characters:".format(number))
    for substring in max_substrings:
        print(substring)

def is_palindrome(string):
    return string == string[::-1]

def print_palindrome(string):
    number = len(string)
    for i in range(number):
        for j in range(i + 1, number + 1):
            substring = string[i:j]
            if is_palindrome(substring):
                print(substring)


string = str(input("Enter a string: "))

print("All possible substrings:")
print_substrings(string)

k = int(input("Enter the length of substrings (K): "))
print(f"All possible substrings of length {k}:")
print_substrings_len_k(string, k)

number = int(input("Enter the length of substrings (K) with N distinct characters: "))
print(f"All substrings of length {number} with {number} distinct characters:")
print_substrings_len_k_distinct(string, number)

d = int(input("Enter the number of distinct characters (N): "))
print_max_length_substrings_distinct(string, d)

print("All palindrome substrings:")
print_palindrome(string)
```
