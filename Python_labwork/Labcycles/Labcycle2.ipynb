```python
#PROGRAM 1

def count_rabbits_pairs(months):
    if months == 1:
        return 1
    elif months == 2:
        return 1
    else:
        rabbits = [1, 1]
        for i in range(2, months):
            rabbits.append(rabbits[i - 1] + rabbits[i - 2])
        return rabbits

def display_table(months):
    print("Month\tPairs of Rabbits")
    pairs = count_rabbits_pairs(months)
    for i in range(months):
        print(f"{i+1}\t\t{pairs[i]}")
num_months = int(input("Enter the number of months: "))
display_table(num_months)
```

```python
#PROGRAM 2

def shift_listr(input_list, k):
    return input_list[-k:] + input_list[:-k]

#Get input and convert it to a list of integers
input_string = input("Enter a string of numbers separated by space: ")
original_list = list(map(int, input_string.split()))

#Rotate elements of the list by 'k' to the right
k = int(input("Enter the number of positions to rotate to the right: "))
rotated_list = shift_listr(original_list, k)

#Convert the list into a tuple using list comprehension and remove duplicates
unique_tuple = tuple({x for x in rotated_list})
unique_list = list(unique_tuple)

#Create another list by putting the results of the function ùëì(ùë•) = ùë•^2 ‚Äì ùë•
function_list = [x ** 2 - x for x in unique_list]

#Sort the two lists individually and merge them to create a single sorted list
sorted_function_list = sorted(function_list)
sorted_rotated_list = sorted(rotated_list)
merged_sorted_list = sorted_rotated_list + sorted_function_list

#Printing all the created lists
print("Original List:", original_list)
print("Rotated List:", rotated_list)
print("Tuple without Duplicates:", unique_tuple)
print("List without Duplicates:", unique_list)
print("Squared Minus x List:", function_list)
print("Sorted Squared Minus x List:", sorted_function_list)
print("Sorted Rotated List:", sorted_rotated_list)
print("Merged Sorted List:", merged_sorted_list)
```

```python
#PROGRAM 3

import json

#create a list having each line of file as element

file = open("iris.json", "r")
txt_file = file.read()
print(txt_file)

def calculate_total_area(flower):
    sepal_area = flower.get("sepalLength") * flower.get("sepalWidth")
    petal_area = flower.get("petalLength") * flower.get("petalWidth")
    return sepal_area + petal_area

# create  a list of dictionary objects

list_of_dictionary = json.loads(txt_file)

# show the details of the flower species naming setosta

for i in list_of_dictionary:
    if i.get("species") == "setosa":
        print(i)

petal_lst = []
sepal_lst = []
for j in list_of_dictionary:
    sepal_length = j.get('sepalLength')
    sepal_width = j.get('sepalWidth')
    petal_length = j.get('petalLength')
    petal_width = j.get('petalWidth')

    # Check for None values before performing calculations
    if all(v is not None for v in [sepal_length, sepal_width, petal_length, petal_width]):
        petal_area = round(petal_length * petal_width, 2)
        petal_lst.append(petal_area)
        sepal_area = round(sepal_length * sepal_width, 2)
        sepal_lst.append(sepal_area)
    else:
        print("Missing values for some attributes in the data.")

# Print the minimum petal area and maximum sepal area in each species

print("MAXIMUM SEPAL AREA:", max(sepal_lst))
print("MINIMUM PETAL AREA:", min(petal_lst))

#Sort the list of dictionaries according to the total area for sepal and petal.

sorted_list = sorted(list_of_dictionary, key=calculate_total_area, reverse=True)
print("Sorted List based on Total Area:")
for item in sorted_list:
    print(item)
```

```python
#PROGRAM 4

class Box:
    def __init__(self, length, breadth=None, height=None):
        # Rectangular prism
        self.Length = length
        self.Breadth = breadth
        self.Height = height

        # Cube
        if breadth is None and height is None:
            self.Breadth = length
            self.Height = length
        elif height is None:
            self.Height = length

    def Area(self):
        area = 2 * (self.Length * self.Breadth + self.Breadth * self.Height + self.Length * self.Height)
        return area

    def volume(self):
        vol = self.Length * self.Breadth * self.Height
        return vol

cube = Box(5)
sq_prism = Box(5, 4)
rect_prism = Box(5, 4, 3)

print("AREA OF CUBE:", cube.Area())
print("VOLUME OF CUBE:", cube.volume())

print("AREA OF SQUARE PRISM:", sq_prism.Area())
print("VOLUME OF SQUARE PRISM:", sq_prism.volume())

print("AREA OF RECTANGULAR PRISM:", rect_prism.Area())
print("VOLUME OF RECTANGULAR PRISM:", rect_prism.volume())
```

```python
#PROGRAM 5

class _3DShapes:
    def printvolume(self):
        pass

    def printarea(self):
        pass

class Cylinder(_3DShapes):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def printvolume(self):
        volume = 3.1416 * (self.radius ** 2) * self.height
        print(f"Volume of Cylinder: {volume:.2f}")

    def printarea(self):
        area = 2 * 3.1416 * self.radius * (self.radius + self.height)
        print(f"Surface Area of Cylinder: {area:.2f}")

class Sphere(_3DShapes):
    def __init__(self, radius):
        self.radius = radius

    def printvolume(self):
        volume = (4 / 3) * 3.1416 * (self.radius ** 3)
        print(f"Volume of Sphere: {volume:.2f}")

    def printarea(self):
        area = 4 * 3.1416 * (self.radius ** 2)
        print(f"Surface Area of Sphere: {area:.2f}")

# Creating instances of Cylinder and Sphere
cylinder = Cylinder(3, 5)
sphere = Sphere(4)

# Printing volume and area of Cylinder and Sphere
cylinder.printvolume()
cylinder.printarea()

sphere.printvolume()
sphere.printarea()
```
